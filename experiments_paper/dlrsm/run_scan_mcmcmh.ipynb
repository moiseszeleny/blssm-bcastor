```python
from hepstack_obj import masses_mH1M2
from hepaid.search.objective import Objective
from hepaid.search.method import MCMCMH

import numpy as np
```

```python
hep_config = "/home/moises/blssm-bcastor/experiments_paper/dlrsm/configs/sphenostack_config.yaml"
obj_hep_config= 'configs/mh1_obj_hep_mcmc.yml'

objective_fn = masses_mH1M2(
        obj_hep_cfg=obj_hep_config,
        hep_stack_cfg=hep_config,
        )

obj = Objective(
    function = objective_fn,
    function_config= obj_hep_config
)
```

```python
mcmcmh = MCMCMH(
    objective=obj,
    hyper_parameters='configs/mcmcmh_config.yaml',
)
```

```python
mcmcmh.run()
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(obj.dataset).astype(float)
```

```python
#data.keys()
```

```python
filtered_data = data.dropna(subset=["mH1", "mH2"])
```

```python
cond_mhi_good = (filtered_data["mH1"] < 130) & (filtered_data["mH1"] > 120) & (filtered_data["mH2"] < 4000) & (filtered_data["mH2"] > 125)
cond_mhi_bad = ~(cond_mhi_good)
selected_data_good = filtered_data[cond_mhi_good]
selected_data_bad = filtered_data[cond_mhi_bad]
```

```python
#selected_data_good
```

```python
combined_data = pd.concat([selected_data_good, selected_data_bad])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(selected_data_bad["lam1input"], selected_data_bad["lam3input"], color='red', label='Bad', alpha=0.3, s=10)
plt.scatter(selected_data_good["lam1input"], selected_data_good["lam3input"], color='blue', label='Good', alpha=0.9, s=20)
plt.xlabel("$\lambda_1$", fontsize=16)
plt.ylabel("$\lambda_3$", fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(selected_data_bad["alpha1input"], selected_data_bad["alpha3input"], color='red', label='Bad', alpha=0.3, s=10)
plt.scatter(selected_data_good["alpha1input"], selected_data_good["alpha3input"], color='blue', label='Good', alpha=0.9, s=20)
plt.xlabel(r"$\alpha_1$", fontsize=16)
plt.ylabel(r"$\alpha_3$", fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(selected_data_bad["mH1"], selected_data_bad["mH2"], color='red', label='Bad', alpha=0.3, s=10)
plt.scatter(selected_data_good["mH1"], selected_data_good["mH2"], color='blue', label='Good', alpha=0.9, s=20)
plt.xlabel("$m_{H_1}$", fontsize=16)
plt.ylabel("$m_{H_2}$", fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend()
plt.show()
```
